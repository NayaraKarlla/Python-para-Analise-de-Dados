## Importando um DataFrame da Web


```python
definição_medalhas = pd.read_csv ('https://raw.githubusercontent.com/letsdata/arquivos-jcd/main/medals.csv')

definição_medalhas.head()
```

```python
definição_medalhas.Sport # forma menos usada para selecionar a coluna inteira sem fazer filtro
```

```python
definição_medalhas ['Sport'] # forma mais usada para selecionar a coluna inteira sem fazer filtro
```

```python
# Loc não é posicional! Ele acesso o índice
definição.loc ['linha1']
```

```python
# Iloc é posicional
definição.iloc [0]
```

```python
# nome_do_dataframe.loc[nome_do_dataframe[nome_da_coluna_para_filtro] == 'critério de filtro]

definição_medalhas.loc[definição_medalhas ['NOC'] == 'FRA']  # aqui o NOC E FRA precisaram ser maiúsculos
```

```python
# Fazendo mais de um filtro
# nome_dataframe.loc[nome_dataframe[nome_coluna_para_filtro1] == 'critério_de_filtro1) &
# (nome_dataframe.loc[nome_dataframe[nome_coluna_para_filtro2] == 'critério_de_filtro2')

definição_medalhas.loc[(definição_medalhas ['NOC'] == 'FRA') & (definição_medalhas ['Medal'] == 'Gold')]
```

```python
# Saber ser o valor está na lista usando o isin, que é usado como 'OU'
definição_medalhas.loc[(definição_medalhas ['NOC'] == 'FRA') & (definição_medalhas ['Medal'].isin(['Gold', 'Silver']))]
```

```python
# Para negar que não está no ouro e não está na prata, é só colocar o ~ na frente da sentença

definição_medalhas.loc[(definição_medalhas ['NOC'] == 'FRA') & (~definição_medalhas ['Medal'].isin(['Gold', 'Silver']))]
```

```python
# Forma para trazer apenas algumas colunas
# nome_dataframe.loc[nome_filtro_linhas, [ coluna1, coluna2, coluna3]]

definição_medalhas.loc[(definição_medalhas ['NOC'] == 'FRA') & (definição_medalhas ['Medal'] == 'Gold'), ['Year', 'Sport', 'Event', 'Medal']]
```

```python
# O Iloc vai selecionar posicionalmente as linhas e as colunas, fazendo um recorte das informações

definição_medalhas.iloc [0:3, 1:3] # 0:3 são as linhas, 1:3 são as colunas na posição 1 e 2
```

```python
# Incluindo mais uma coluna na busca
# O Iloc vai selecionar posicionalmente as linhas e as colunas, fazendo um recorte das informações

definição_medalhas.iloc [0:3, 1:4]  # buscou a coluna "discipline"
```

```python
# É possivel criar diferente filtros, o pipe '|' é o 'OU', e o '&' é o E (and)

definição_medalhas.loc[(definição_medalhas ['NOC'] == 'FRA') | (definição_medalhas ['NOC'] == 'AUT')]
```


##JSON

Vamos aprender um pouco sobre manipulação do formato JSON. Arquivos de configuração, requisições e respostas de APIs  e tantas outras aplicações desse formato.

O Json se parece muito com os dicionários, temos de graça, sem necessidade de pacotes adicionais, com uma boa estruturação para ler, escrever e manipular esses tipos de arquivos.

Uma ótima referência é a própria [documentação do Python](https://docs.python.org/3/library/json.html), a seguir vamos manipular arquivos JSON com nosso pacote de manipulação favorito!

```python
import pandas as pd
import json
```

Mas Pandas não trabalha com dados tabulares? Sim, verdade, mas podemos fazer algumas transformações para deixar os dados tabulares, ou mesmo dizer para o Pandas como a gente quer que ele interprete os dados estruturados.

No método read_json existe um parâmetro "orient" que possui as alternativas de valor:

- 'split' : dicionarios com formato {index -> [indice], columns -> [colunas], data -> [valores]}

- 'records' : listas com formato [{coluna -> valor}, ... , {coluna -> valor}]

- 'index' : dicionários com formato {indice -> {coluna -> valor}}

- 'columns' : dicionarios com formato {coluna -> {indice -> valor}}

- 'values' : um arranjo com valores

```python


dicionario_com_valores = {
    "data": [[2,120],[3,1],[1,100],[1,100], [1, 10], [3, 8], [2, 80]],
    "index": ["hamburguer","alface","queijo","molho especial", "cebola", "picles", "pão com gergelim"],
    "columns": ["qtd","calorias"]
}

dicionario_com_valores
```

```python
dicionario_com_valores.keys()
```

```python
type(dicionario_com_valores)
```

A função dumps é usada para converter objetos Python em strings no formato JSON. Ela faz parte do módulo json da biblioteca padrão.

O Python chama o json de dicionário por ser exatamente a mesma estrutura.

```python
str_valores = json.dumps(dicionario_com_valores) # transformou um dict(dicionário) em string no formato json

str_valores
```

```python
type(json.dumps(dicionario_com_valores)) # formato string
```

```python
type (str_valores)
```

```python
json.loads (str_valores) # loads transforma uma string json em dict = dicionário
```

```python
# str_valores será colocada dentro do Pandas

str_json_split = json.dumps({
     "data": [[2,120],[3,1],[1,100],[1,100], [1, 10], [3, 8], [2, 80]],
    "index": ["hamburguer","alface","queijo","molho especial", "cebola", "picles", "pão com gergelim"],
    "columns": ["qtd","calorias"]
})

str_json_split
```

```python
# Com o orient split o Pandas espera três chaves: index, columns e data
# O oriente é o parâmetro de como o Pandas deve tratar os dados que são passados para ele

# Lendo a string JSON e convertendo para DataFrame

pd.read_json(str_json_split, orient = 'split')
```

```python
# registros: são listas com formato [{coluna:valor}, ..., {coluna:valor }]

json_registros = json.dumps ([
    {"ingrediente":"hamburguer","qtd":2,"calorias":120}, # ingrediente é uma coluna e hamburguer é um valor, e assim por diante...essa é a primeira linha, é um dicionárioa, juntando com as outras linhas forma-se uma conjuntos decionários
    {"ingrediente":"alface","qtd":3,"calorias":1},
    {"ingrediente":"queijo","qtd":1,"calorias":100},
    {"ingrediente":"molho especial","qtd":1,"calorias":100}
])

json_registros # retornar uma lista de dicionários
```

```python
pd.read_json(json_registros, orient = 'records') # estou dizendo a forma como o Pandas vai receber o json, que no caso é em formato de record, que são registros
```

```python
# index: dicionários com formato {indice (é a chave) -> {coluna -> valor}}
# o indice é a chave e o valor são as colunas com os dados!
# esse é um formato totalmente diferente

json_index = json.dumps({
    "hamburguer": {
        "qtd": 5,
        "calorias": 20
        },
    "alface": {
        "qtd": 8,
        "calorias": 30
        },
    "queijo": {
        "qtd": 2,
        "calorias": 120
        },
    "molho especial": {
        "qtd": 4,
        "calorias": 16
        }

})

json_index
```

```python
# lendo o JSON
pd.read_json(json_index, orient = 'index')
```

```python
# columns: dicionarios com formato {coluna -> {indice -> valor}}

# aqui a chave é a COLUNA e os valores são pares indice_pandas:valor_coluna

json_columns = json.dumps({
    "qtd": {
        "hamburguer": 2,
        "alface": 1,
        "queijo": 1,
        "molho especial": 1
    },
    "calorias": {
        "hamburguer": 120,
        "alface": 1,
        "queijo": 100,
        "molho especial": 100
    }
})

json_columns
```

```python
pd.read_json(json_columns, orient = 'columns')
```

```python
# 'values' : um arranjo com valores, cada linha será uma lista

json_values = json.dumps([
    ["hamburguer", 2, 120],
    ["alface", 1, 1],
    ["queijo", 1,100],
    ["molho especial", 1, 100]

])

json_values
```

```python
# aqui ele vai entender que cada linha são valores, só que o nome das colunas não foram informados por isso ficou registrado como 0,1 e 2
pd.read_json(json_values, orient = 'values')
```

## Estruturando dados hierarquicos em formato tabular normalizando os mesmos

```python
# Uma forma legal de trabalhar com dados mais aninhados é utilizar o json_normalize do Pandas.

# Para entender como normalizar (achatar) o json é preciso pensar ANTES como queremos o resultado.

# No caso das bandas, faz sentido ter uma linha para cada música, com as colunas para os nomes das bandas e dos álbuns. Sabendo qual o nível mais "profundo" que queremos ir, fica mais fácil.

# Recuperando a estrutura de bandas original:

with open('bands.json') as arquivo_json:
    objeto_json = json.load(arquivo_json)

objeto_json
```

```python
# 'pd.json_normalize' do Pandas, serve para transformar estruturas de JSON aninhado em um DataFrame plano.

# O objeto_JSON que contém os dados aninhados

# record_path=['albums', 'songs']: Indicação para descer dois níveis no JSON: primeiro entra em albums e depois em songs

# meta=['name', ['albums', 'title']]: Esse parâmetro especifica os dados que serão mantidos em cada linha: 'name" que é o nome da banda e ['albums', 'title'] é o título do álbum onde a música está


definição_bandas = pd.json_normalize(objeto_json, record_path=['albums', 'songs'], meta=['name', ['albums', 'title']])

definição_bandas
```
